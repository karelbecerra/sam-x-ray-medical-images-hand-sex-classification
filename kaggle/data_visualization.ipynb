```python
import numpy as np
import matplotlib.pyplot as plt

def display_sample(img_array, n_img):
  n_img = min(n_img, img_array.shape[0])
  fig, ax = plt.subplots(1, n_img, figsize=(n_img*2, 2))
  for i in range(n_img):
      ax[i].imshow(img_array[i], cmap='gray')
  plt.show()
```
